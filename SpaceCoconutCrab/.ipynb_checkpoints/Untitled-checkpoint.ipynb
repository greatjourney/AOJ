```python
import pandas as pd
import matplotlib
matplotlib.use('Agg') 
import matplotlib.pyplot as plt

df1 = pd.read_csv('zy_out.csv')
df2 = pd.read_csv('xy_out.csv')

plt.plot('e', 'cnt', data=df1, label='zy', marker='o')
plt.plot('e', 'cnt', data=df2, label='xy', marker='o')
plt.xscale('log')
plt.yscale('log')
plt.xlabel("E")
plt.ylabel("count")
plt.savefig('figure.png')

        
        
    
```
